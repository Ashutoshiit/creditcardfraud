```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['card_transdata.csv']))
print(df)
```

```python
import tensorflow as tf 
import numpy as np
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt 
import plotly.express as px
corr = df.corr()
px.imshow(corr)
```

```python
import keras 
model = tf.keras.Sequential([
    tf.keras.layers.Dense(units = 256, activation = 'relu', input_shape = (7,)),
    tf.keras.layers.Dense(units = 1024, activation = 'relu'),
    tf.keras.layers.Dense(units = 1, activation = 'sigmoid')]
)
model.compile(loss=tf.keras.losses.BinaryCrossentropy(), 
         optimizer='adam',
         metrics=['accuracy'])

```

```python
print(model.summary())
```

```python
df['distance_from_home'] /= max(df['distance_from_home'])
df['distance_from_last_transaction'] /= max(df['distance_from_last_transaction'])	
df['ratio_to_median_purchase_price'] /= max(df['ratio_to_median_purchase_price'])
df.describe()
```

```python
N = 7
# Select first N columns
first_n_column  = df.iloc[: , :N]
print(first_n_column)
```

```python
model.fit(first_n_column,df['fraud'],epochs = 5)
```
